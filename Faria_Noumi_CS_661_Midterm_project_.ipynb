```python
# Check for outliers using IQR
q1 = df1.quantile(0.25)
q3 = df1.quantile(0.75)
IQR = q3 - q1

not_outliers = df1[~((df1 < (q1 - 1.5 * IQR)) | (df1 > (q3 + 1.5 * IQR)))]

# Outliers filled with zeros.
New_df1 = not_outliers.fillna(value=np.nan)


# Display the first 10 rows of the new DataFrame
New_df1.head(10)

```

```python
# Create a boxplot using seaborn
sns.boxplot(data=New_df1, color='blue')

ax = plt.gca() #Retrieve the axis object:
labels = ax.get_xticklabels() #Get the x-axis tick labels:
plt.setp(labels, rotation=90, horizontalalignment='center') #Rotate the labels and set the horizontal alignment:

plt.show()
```

```python
#*args syntax allows you to write functions that can accept a flexible number of non-keyword arguments.
#The **kwargs syntax allows you to accept a variable number of keyword arguments in your Python function.
def function(*args, **kwargs):
    # Access non-keyword arguments using args
    for arg in args:
        print("Non-keyword argument:", arg)

    # Access keyword arguments using kwargs
    for key, value in kwargs.items():
        print("Keyword argument:", key, "=", value)
```
